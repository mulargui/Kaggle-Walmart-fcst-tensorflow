```
import pandas as pd

print(pd.__version__)
```

```
#load data
dftrain=pd.read_csv('https://raw.githubusercontent.com/mulargui/Kaggle-Walmart-fcst-tensorflow/master/kaggle/train.csv')
dftest=pd.read_csv('https://raw.githubusercontent.com/mulargui/Kaggle-Walmart-fcst-tensorflow/master/kaggle/test.csv')

print(dftrain.shape[0])
print(dftest.shape[0])
```

```
#concat both dfs and get all dates
dfresult=pd.concat([dftrain,dftest])

print(dfresult.shape[0])
```

```
#leave only Date and IsHoliday
dfresult.drop("Dept", axis=1, inplace=True) 
dfresult.drop("Store", axis=1, inplace=True) 
dfresult.drop("Weekly_Sales", axis=1, inplace=True) 

print(dfresult.shape[0])
```

```
#eliminate duplicates, this is the final list of Dates to work with
dfresult.drop_duplicates(inplace = True)

print(dfresult.shape[0])
```

```
#Add Seasonality Column based on Week of the Date
dfresult['Seasonality'] = pd.to_datetime(dfresult['Date']).apply(lambda x: x.weekofyear)
```

```
#Add Year column based on Date
dfresult['Year'] = pd.to_datetime(dfresult['Date']).apply(lambda x: x.year)
```
