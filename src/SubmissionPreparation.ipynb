```
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```
#load data
from google.colab import drive
drive.mount('/content/drive/')

dftest=pd.read_csv('/content/drive/My Drive/test.csv')
dfpredict=pd.read_csv('/content/drive/My Drive/prediction.csv', header=None)

#rename the prediction column
dfpredict.rename(columns={dfpredict.columns[0]: 'Weekly_Sales'}, inplace=True)

#All data together
dftest = dftest.join(dfpredict)

# Add the Id column
dftest['Id'] = dftest['Store'].map(str) + '_' + dftest['Dept'].map(str) + '_' + dftest['Date'].map(str)

#Round the prediction to 2 decimals (cents)
dftest['Weekly_Sales'] = dftest['Weekly_Sales'].astype(np.float64)
dftest['Weekly_Sales'] = round(dftest['Weekly_Sales'],2)

#Select the attributes needed to submit
dfsolution = dftest[['Id', 'Weekly_Sales']].copy()

print(dfsolution.head())
print(dfsolution.shape[0])
```

```
#Save the dataframe in a csv file in Google Drive
from google.colab import drive
drive.mount('/content/drive/')
dfsolution.to_csv('/content/drive/My Drive/Submission.csv', na_rep='NA', index=False)
```
