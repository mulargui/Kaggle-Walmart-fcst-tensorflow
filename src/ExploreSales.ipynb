```
#used to explore sales data

import pandas as pd

print(pd.__version__)
```

```
#load data
dftrain=pd.read_csv('https://raw.githubusercontent.com/mulargui/Kaggle-Walmart-fcst-tensorflow/master/kaggle/train.csv')

print(dftrain.shape[0])
```

```
##########################################################
# This section generates Date features
##########################################################
#Add Seasonality Column based on Week of the Date
dftrain['Seasonality'] = pd.to_datetime(dftrain['Date']).apply(lambda x: x.weekofyear)
#Add Year column based on Date
dftrain['Year'] = pd.to_datetime(dftrain['Date']).apply(lambda x: x.year)

print(dftrain.shape[0])
```

```
#explore sales in weeks 49,50,51,52
dfchristmas = dftrain[(dftrain.Year == 2011) & (dftrain.Seasonality > 47)]

print(dfchristmas.shape[0])
```

```
#we only adjust those with all weeks of sales
dfchristmas = dfchristmas.groupby(['Store', 'Dept']).filter(lambda x: len(x) == 5 )

print(dfchristmas.shape[0])
```

```
#consolidate sales per store
dfchristmas = dfchristmas.groupby(['Store', 'Seasonality'])['Weekly_Sales'].sum().unstack()
dfchristmas = pd.DataFrame(dfchristmas.to_records())

print(dfchristmas.shape[0])
```

```
#explore sales increment week over week
dfchristmas['49inc'] = dfchristmas['49'] / dfchristmas['48']
dfchristmas['50inc'] = dfchristmas['50'] / dfchristmas['49']
dfchristmas['51inc'] = dfchristmas['51'] / dfchristmas['50']
dfchristmas['52inc'] = dfchristmas['52'] / dfchristmas['51']

print(dfchristmas.shape[0])
print (dfchristmas)
```
